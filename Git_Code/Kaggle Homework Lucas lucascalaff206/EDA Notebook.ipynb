# EDA Notebook

```python
import pandas as pd
import numpy as np
import sweetviz as sv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Irrigation = pd.read_csv('train.csv')
report = sv.analyze(Irrigation)
report.show_html('Irrigation_EDA_Report.html')
```

```python
# Looking at the mean values and standard deviations of soil ph by soil type
soil_ph_by_type = Irrigation.groupby('Soil_Type')['Soil_pH'].agg(['mean', 'std'])
print(soil_ph_by_type)
```

```python
# Looking into soil moisture levels by soil type
soil_moisture_by_type = Irrigation.groupby('Soil_Type')['Soil_Moisture'].agg(['mean', 'std'])
print(soil_moisture_by_type)
```

```python
# Looking into temprature levels by soil type
temperature_by_type = Irrigation.groupby('Soil_Type')['Temperature_C'].agg(['mean', 'std'])
print(temperature_by_type)
```

```python
# value count of regions 
print(Irrigation['Region'].value_counts())
```

```python
# Showing how each regions soil type is distributed
plt.figure(figsize=(10, 6))
sns.countplot(data=Irrigation, x='Region', hue='Soil_Type')
plt.title('Distribution of Soil Types by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='Soil Type')
plt.show()
```

```python
# look at irrigation need by soil type, irrigation is categorical so we can look at the distribution of irrigation needs by soil type
plt.figure(figsize=(10, 6))
sns.countplot(data=Irrigation, x='Soil_Type', hue='Irrigation_Need')
plt.title('Distribution of Irrigation Needs by Soil Type')
plt.xlabel('Soil Type')
plt.ylabel('Count')
plt.legend(title='Irrigation Need')
plt.show()
```

```python
# Look at irrigation need by region 
plt.figure(figsize=(10, 6))
sns.countplot(data=Irrigation, x='Region', hue='Irrigation_Need')
plt.title('Distribution of Irrigation Needs by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='Irrigation Need')
plt.show()
```
