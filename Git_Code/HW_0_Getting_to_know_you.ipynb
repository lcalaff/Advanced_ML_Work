## Part 2 - Cost of Living and Salary 

```python
import pandas as pd 
```

https://www.kaggle.com/datasets/mvieira101/global-cost-of-living/data

^^ Link to Data i used 

```python
# Cost of Living Data Set was taken 3 years ago so may not accuratly reflect todays prices
# Monthly Pricing for Variables (Will change names to improve readability)
df = pd.read_csv("cost-of-living.csv")
```

```python
df.rename(columns = {"x1":"Meal", "x12": "Eggs", "x33":"Gas_L","x36": "Utilitys", "x37":"Phone_Plan",
"x38":"Internet", "x39":"Gym", "x49":"Apartment", "x54": "Salary"}, inplace = True)
```

```python
Sea_df = df[df["city"] == "Seattle"]
Sdo_df = df[df["city"] == "San Diego"]
LA_df = df[df["city"] == "Los Angeles"]
```

Calculation Functions 

```python
def housing_price(df):
    P = df[["Utilitys", "Apartment", "Internet", "Phone_Plan", "Gym"]].to_numpy().sum()
    return P
```

```python
def Food_Price(df):
    return (df["Meal"] * 5 + 500).sum()
    # Assuming mt monthly gorcieries are maximum $500
    # Also asumming that you eat out once a week 
```

```python
def Transportation_Price(df):
    return (df["Gas_L"] * 207 + 50) .sum()
    # 207: Avg Liters a US driver uses per month 
    # $50 Goes towards Paying for parking which could vary significantly (i personally try to 
    # almost never pay to park)
```

Housing Calculations per state

```python
Housing_Sea = housing_price(Sea_df)
Housing_Sdo = housing_price(Sdo_df)
Housing_LA = housing_price(LA_df)
```

Food Calculations per state 

```python
Food_Sea = Food_Price(Sea_df)
Food_Sdo = Food_Price(Sdo_df)
Food_LA = Food_Price(LA_df)
```

Transportation Calculations per State 

```python
Car_Sea = Transportation_Price(Sea_df)
Car_Sdo = Transportation_Price(Sdo_df)
Car_La = Transportation_Price(LA_df)
```

minimum Salary Needed 

```python
min_Sal_Sea = Housing_Sea + Food_Sea + Car_Sea
min_Sal_Sdo = Housing_Sdo + Food_Sdo + Car_Sdo
min_Sal_LA = Housing_LA + Food_LA + Car_La
```

Desired Salary 

```python
Data_Analyst_Sal = 75000/12
```

Living Costs DataFrame 

```python
Liv_Costs = pd.DataFrame({
    "City" : ["Seattle", "San Diego", "Los Angeles"], 
    "Housing" : [Housing_Sea, Housing_Sdo, Housing_LA], 
    "Food" : [Food_Sea, Food_Sdo, Food_LA], 
    "Transportation" : [Car_Sea, Car_Sdo, Car_La],
    "Minimum_Salary" : [min_Sal_Sea, min_Sal_Sdo, min_Sal_LA], 
    "Desired_Salary" : [Data_Analyst_Sal, Data_Analyst_Sal, Data_Analyst_Sal],
    "Min_Wage": [3692, 3076, 3097]
})
```

Plotting Min Wage vs Monthly Living Costs 

```python
Liv_Costs.plot(
    x = "City", 
    y = ["Min_Wage", "Minimum_Salary"], 
    kind = "bar"
)
```
